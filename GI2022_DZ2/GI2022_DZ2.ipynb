```python
import pysam
import os
print(os.getcwd())

pysam.index("drive/MyDrive/merged-tumor.bam")
bamfile = pysam.AlignmentFile("drive/MyDrive/merged-tumor.bam", "rb")
```

```python
bamfile = pysam.AlignmentFile("drive/MyDrive/merged-tumor.bam", "rb")
for read in bamfile:
  print(f'FLAG: {read.flag}')
  print(f'MAPPING QUALITY: {read.mapping_quality}')
  break
```

```python
bamfile = pysam.AlignmentFile("drive/MyDrive/merged-tumor.bam", "rb")
print(f'Number of mapped reads: {bamfile.mapped}')
print(f'Number of unmapped reads: {bamfile.unmapped}')
print(f'Number of reads total: {bamfile.unmapped + bamfile.mapped}')
```

```python
bamfile = pysam.AlignmentFile("drive/MyDrive/merged-tumor.bam", "rb")
countmq0 = 0
for read in bamfile.fetch(until_eof=True):
  if read.mapping_quality == 0:
    countmq0 += 1
print(f'Number of reads with mapping quality 0: {countmq0}')
```

```python
bamfile = pysam.AlignmentFile("drive/MyDrive/merged-tumor.bam", "rb")
sum = 0.0
for read in bamfile.fetch(until_eof=True):
  sum += read.mapping_quality
print(sum)
print(bamfile.count())
print(f'Average mapping quality for all the reads: { sum / bamfile.count()}')
```

```python
bamfile = pysam.AlignmentFile("drive/MyDrive/merged-tumor.bam", "rb")
sum = 0.0
cnt = 0
for read in bamfile.fetch(until_eof=True):
  if read.mapping_quality != 0:
    sum += read.mapping_quality
    cnt += 1

print(f'Average mapping quality for all the reads with mapping quality > 0: { sum / cnt}')
```
